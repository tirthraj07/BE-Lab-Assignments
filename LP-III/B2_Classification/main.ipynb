## Assignment B2
Classify the email using the binary classification method. Email Spam detection has two states: 
a) Normal State Not Spam 
b) Abnormal State Spam. Use K-Nearest Neighbors and Support Vector Machine for classification. Analyze the performance. 

```python
import pandas as pd
```

```python
df = pd.read_csv('emails.csv')
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
df.drop(['Email No.'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['Prediction'], axis=1)
y = df['Prediction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("KNN Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("\n\nKNN Classification Report: \n", classification_report(y_test, y_pred_knn))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', probability=True)
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
y_pred_svm
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
print("Classification Report: \n", classification_report(y_test, y_pred_svm))
```
