Implement Gradient Descent Algorithm to find the local minima of a function. 
For example, find the local minima of the function y=(x+3)Â² starting from the 
point x=2

```python
# Define the function y = (x + 3)^2
def func(x):
    return (x + 3)**2
```

```python
# Define the derivative of the function: dy/dx = 2(x + 3)
def derivative(x):
    return 2 * (x + 3)
```

```python
gd = []

def gradient_descent(start, learning_rate, max_iter):
    x = start
    gd.append(x)
    for i in range(max_iter):
        grad = derivative(x)
        x -= learning_rate * grad
        y = func(x)
        gd.append(x)
        print(f"Iteration {i + 1}, x = {x}, f(x) = {y}")
    
    return x
```

```python
start = 3
learning_rate = 0.01
max_iter = 200

min_x = gradient_descent(start, learning_rate, max_iter)
print(f"Local Minima occurs at x = {min_x}")
```
